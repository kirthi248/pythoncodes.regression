```python
user_movie = movie.pivot(index='userId',
                                 columns='movie',
                                 values='rating')
```

```python
#user_movie.index = movie.userId.unique()
```

```python
user_movie
```

```python
#Impute those NaNs with 0 values
user_movie.fillna(0, inplace=True)
```

```python
user_movie
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_movie.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = movie.userId.unique()
user_sim_df.columns = movie.userId.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
